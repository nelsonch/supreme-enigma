```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (16, 10)
plt.rcParams['font.size'] = 11

# Define saturation functions
def hill_saturation(x, K, S):
    """Hill saturation function"""
    return x**S / (K**S + x**S)

def weibull_saturation(x, K, S):
    """Weibull saturation function"""
    return 1 - np.exp(-(x/K)**S)

# Generate x values
x = np.linspace(0, 10000, 1000)

# Create figure with 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# ============================================================================
# 1. Hill C-shape (S > 1): Convex, slow start then accelerating
# ============================================================================
ax = axes[0, 0]
K_fixed = 5000
S_values = [1.5, 2.0, 2.5, 3.0]
colors = plt.cm.Blues(np.linspace(0.4, 0.9, len(S_values)))

for S, color in zip(S_values, colors):
    y = hill_saturation(x, K_fixed, S)
    ax.plot(x, y, linewidth=2.5, label=f'S={S}', color=color)

ax.set_title('Hill C-shape: $f(x) = \\frac{x^S}{K^S + x^S}$ with S > 1 (Convex)', 
             fontsize=14, fontweight='bold')
ax.set_xlabel('Marketing Spend', fontsize=12)
ax.set_ylabel('Response (Saturation)', fontsize=12)
ax.legend(title=f'Shape (K={K_fixed})', fontsize=10, title_fontsize=11)
ax.grid(True, alpha=0.3)
ax.set_ylim([0, 1.05])
ax.text(0.05, 0.95, 'Convex: Accelerating returns\nthen saturation', 
        transform=ax.transAxes, fontsize=11, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.3))

# ============================================================================
# 2. Hill S-shape (0 < S < 1): S-shaped, slow-fast-slow
# ============================================================================
ax = axes[0, 1]
K_fixed = 5000
S_values = [0.3, 0.5, 0.7, 0.9]
colors = plt.cm.Greens(np.linspace(0.4, 0.9, len(S_values)))

for S, color in zip(S_values, colors):
    y = hill_saturation(x, K_fixed, S)
    ax.plot(x, y, linewidth=2.5, label=f'S={S}', color=color)

ax.set_title('Hill S-shape: $f(x) = \\frac{x^S}{K^S + x^S}$ with 0 < S < 1 (S-curve)', 
             fontsize=14, fontweight='bold')
ax.set_xlabel('Marketing Spend', fontsize=12)
ax.set_ylabel('Response (Saturation)', fontsize=12)
ax.legend(title=f'Shape (K={K_fixed})', fontsize=10, title_fontsize=11)
ax.grid(True, alpha=0.3)
ax.set_ylim([0, 1.05])
ax.text(0.05, 0.95, 'S-shaped: Initial slow,\nthen fast, then saturation', 
        transform=ax.transAxes, fontsize=11, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.3))

# ============================================================================
# 3. Weibull C-shape (0 < S < 1): Concave, fast start then diminishing
# ============================================================================
ax = axes[1, 0]
K_fixed = 5000
S_values = [0.3, 0.5, 0.7, 0.9]
colors = plt.cm.Oranges(np.linspace(0.4, 0.9, len(S_values)))

for S, color in zip(S_values, colors):
    y = weibull_saturation(x, K_fixed, S)
    ax.plot(x, y, linewidth=2.5, label=f'S={S}', color=color)

ax.set_title('Weibull C-shape: $f(x) = 1 - e^{-(x/K)^S}$ with 0 < S < 1 (Concave)', 
             fontsize=14, fontweight='bold')
ax.set_xlabel('Marketing Spend', fontsize=12)
ax.set_ylabel('Response (Saturation)', fontsize=12)
ax.legend(title=f'Shape (K={K_fixed})', fontsize=10, title_fontsize=11)
ax.grid(True, alpha=0.3)
ax.set_ylim([0, 1.05])
ax.text(0.05, 0.95, 'Concave: Strong initial returns\nthen diminishing', 
        transform=ax.transAxes, fontsize=11, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='bisque', alpha=0.3))

# ============================================================================
# 4. Weibull S-shape (S > 1): S-shaped, similar to Hill S but different inflection
# ============================================================================
ax = axes[1, 1]
K_fixed = 5000
S_values = [1.5, 2.0, 2.5, 3.0]
colors = plt.cm.Purples(np.linspace(0.4, 0.9, len(S_values)))

for S, color in zip(S_values, colors):
    y = weibull_saturation(x, K_fixed, S)
    ax.plot(x, y, linewidth=2.5, label=f'S={S}', color=color)

ax.set_title('Weibull S-shape: $f(x) = 1 - e^{-(x/K)^S}$ with S > 1 (S-curve)', 
             fontsize=14, fontweight='bold')
ax.set_xlabel('Marketing Spend', fontsize=12)
ax.set_ylabel('Response (Saturation)', fontsize=12)
ax.legend(title=f'Shape (K={K_fixed})', fontsize=10, title_fontsize=11)
ax.grid(True, alpha=0.3)
ax.set_ylim([0, 1.05])
ax.text(0.05, 0.95, 'S-shaped: Slow start,\nrapid growth, then saturation', 
        transform=ax.transAxes, fontsize=11, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='plum', alpha=0.3))

plt.tight_layout()
plt.savefig('saturation_functions_4types.png', dpi=300, bbox_inches='tight')
plt.show()

print("Figure saved as 'saturation_functions_4types.png'")

# ============================================================================
# Additional visualization: Varying K parameter (scale)
# ============================================================================

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# ============================================================================
# 1. Hill C-shape: Varying K
# ============================================================================
ax = axes[0, 0]
S_fixed = 2.0
K_values = [3000, 5000, 7000, 9000]
colors = plt.cm.Blues(np.linspace(0.4, 0.9, len(K_values)))

for K, color in zip(K_values, colors):
    y = hill_saturation(x, K, S_fixed)
    ax.plot(x, y, linewidth=2.5, label=f'K={K}', color=color)

ax.set_title('Hill C-shape: Varying Scale Parameter K (S=2.0 fixed)', 
             fontsize=14, fontweight='bold')
ax.set_xlabel('Marketing Spend', fontsize=12)
ax.set_ylabel('Response (Saturation)', fontsize=12)
ax.legend(title=f'Scale (S={S_fixed})', fontsize=10, title_fontsize=11)
ax.grid(True, alpha=0.3)
ax.set_ylim([0, 1.05])

# ============================================================================
# 2. Hill S-shape: Varying K
# ============================================================================
ax = axes[0, 1]
S_fixed = 0.5
K_values = [3000, 5000, 7000, 9000]
colors = plt.cm.Greens(np.linspace(0.4, 0.9, len(K_values)))

for K, color in zip(K_values, colors):
    y = hill_saturation(x, K, S_fixed)
    ax.plot(x, y, linewidth=2.5, label=f'K={K}', color=color)

ax.set_title('Hill S-shape: Varying Scale Parameter K (S=0.5 fixed)', 
             fontsize=14, fontweight='bold')
ax.set_xlabel('Marketing Spend', fontsize=12)
ax.set_ylabel('Response (Saturation)', fontsize=12)
ax.legend(title=f'Scale (S={S_fixed})', fontsize=10, title_fontsize=11)
ax.grid(True, alpha=0.3)
ax.set_ylim([0, 1.05])

# ============================================================================
# 3. Weibull C-shape: Varying K
# ============================================================================
ax = axes[1, 0]
S_fixed = 0.5
K_values = [3000, 5000, 7000, 9000]
colors = plt.cm.Oranges(np.linspace(0.4, 0.9, len(K_values)))

for K, color in zip(K_values, colors):
    y = weibull_saturation(x, K, S_fixed)
    ax.plot(x, y, linewidth=2.5, label=f'K={K}', color=color)

ax.set_title('Weibull C-shape: Varying Scale Parameter K (S=0.5 fixed)', 
             fontsize=14, fontweight='bold')
ax.set_xlabel('Marketing Spend', fontsize=12)
ax.set_ylabel('Response (Saturation)', fontsize=12)
ax.legend(title=f'Scale (S={S_fixed})', fontsize=10, title_fontsize=11)
ax.grid(True, alpha=0.3)
ax.set_ylim([0, 1.05])

# ============================================================================
# 4. Weibull S-shape: Varying K
# ============================================================================
ax = axes[1, 1]
S_fixed = 2.0
K_values = [3000, 5000, 7000, 9000]
colors = plt.cm.Purples(np.linspace(0.4, 0.9, len(K_values)))

for K, color in zip(K_values, colors):
    y = weibull_saturation(x, K, S_fixed)
    ax.plot(x, y, linewidth=2.5, label=f'K={K}', color=color)

ax.set_title('Weibull S-shape: Varying Scale Parameter K (S=2.0 fixed)', 
             fontsize=14, fontweight='bold')
ax.set_xlabel('Marketing Spend', fontsize=12)
ax.set_ylabel('Response (Saturation)', fontsize=12)
ax.legend(title=f'Scale (S={S_fixed})', fontsize=10, title_fontsize=11)
ax.grid(True, alpha=0.3)
ax.set_ylim([0, 1.05])

plt.tight_layout()
plt.savefig('saturation_functions_varying_K.png', dpi=300, bbox_inches='tight')
plt.show()

print("Figure saved as 'saturation_functions_varying_K.png'")

# ============================================================================
# Summary table of characteristics
# ============================================================================

print("\n" + "="*80)
print("SATURATION FUNCTION CHARACTERISTICS")
print("="*80)
print("\n1. Hill C-shape: f(x) = x^S / (K^S + x^S) with S > 1")
print("   - Shape: Convex (C-shaped)")
print("   - Behavior: Accelerating returns initially, then saturation")
print("   - S parameter: Controls convexity (larger S = more convex)")
print("   - K parameter: Half-saturation point (where f(K) = 0.5)")

print("\n2. Hill S-shape: f(x) = x^S / (K^S + x^S) with 0 < S < 1")
print("   - Shape: S-shaped (sigmoid-like)")
print("   - Behavior: Slow start, rapid growth, then saturation")
print("   - S parameter: Controls S-curve steepness (smaller S = more pronounced S)")
print("   - K parameter: Inflection point location")

print("\n3. Weibull C-shape: f(x) = 1 - exp(-(x/K)^S) with 0 < S < 1")
print("   - Shape: Concave (inverted C-shaped)")
print("   - Behavior: Strong initial returns, then diminishing")
print("   - S parameter: Controls concavity (smaller S = more concave)")
print("   - K parameter: Scale parameter (63.2% saturation point)")

print("\n4. Weibull S-shape: f(x) = 1 - exp(-(x/K)^S) with S > 1")
print("   - Shape: S-shaped (different from Hill S)")
print("   - Behavior: Slow initial, accelerating, then saturation")
print("   - S parameter: Controls S-curve shape (larger S = sharper transition)")
print("   - K parameter: Controls location of inflection")

print("\n" + "="*80)
```
